```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tickers= ['QQQ','BIL','VGIT','VCIT','HYG','EPU']
start_date = '2019-12-31'
end_date = '2024-06-17'
```

```python
datos=yf.download(tickers=tickers,start=start_date,end=end_date)['Adj Close']
datos
```

```python
retornos=datos.pct_change().dropna()*100
retornos
```

```python
retornos_acum=datos.pct_change().cumsum().dropna()*100
retornos_acum
```

```python
plt.figure(figsize=(15,6))
plt.style.use('ggplot')
plt.plot(retornos_acum.index,retornos_acum['QQQ'],label='Invesco QQQ Trust Series 1')
plt.plot(retornos_acum.index,retornos_acum['BIL'],label='SPDR Bloomberg 1-3 Month T-Bill ETF')
plt.plot(retornos_acum.index,retornos_acum['VGIT'],label='Vanguard Intermediate-Term Treasury ETF')
plt.plot(retornos_acum.index,retornos_acum['VCIT'],label='Vanguard Intermediate-Term Corp Bond ETF')
plt.plot(retornos_acum.index,retornos_acum['HYG'],label='iShares iBoxx $ High Yield Corporate Bond ETF')
plt.plot(retornos_acum.index,retornos_acum['EPU'],label='iShares MSCI Peru and Global Exposure ETF')
plt.text(retornos_acum.index[-1], retornos_acum['QQQ'][-1],s=round(retornos_acum['QQQ'][-1],1))
plt.text(retornos_acum.index[-1], retornos_acum['BIL'][-1],s=round(retornos_acum['BIL'][-1],1))
plt.text(retornos_acum.index[-1], retornos_acum['VGIT'][-1],s=round(retornos_acum['VGIT'][-1],1))
plt.text(retornos_acum.index[-1], retornos_acum['VCIT'][-1],s=round(retornos_acum['VCIT'][-1],1))
plt.text(retornos_acum.index[-1], retornos_acum['HYG'][-1],s=round(retornos_acum['HYG'][-1],1))
plt.text(retornos_acum.index[-1], retornos_acum['EPU'][-1],s=round(retornos_acum['EPU'][-1],1))
plt.legend()
plt.title('Evolución de los Principales ETFs')
plt.xlabel('Tiempo')
plt.ylabel('Retorno % Acumulado')
plt.axhline(y=0,color='black')
plt.show
```
